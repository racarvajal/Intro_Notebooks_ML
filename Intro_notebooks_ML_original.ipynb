Many magnitudes in this catalogue are in the Vega system.  
We will transform them into the AB system.

For Vega WISE magnitudes, we have the following conversion to AB values.

Conversion to the AB system (from [WISE documentation](https://wise2.ipac.caltech.edu/docs/release/allsky/expsup/sec4_4h.html),  
and [Jarrett et al., 2017](https://ui.adsabs.harvard.edu/abs/2017ApJ...836..182J/abstract))

($m_{\mathrm{AB}} = m_{\mathrm{Vega}} + \Delta (m)$)

|Band | magnitude offset (Δm)|
|-----|---------------------:|
|W1   |                 2.699|
|W2   |                 3.339|
|W3   |                 5.174|
|W4   |                 6.620|

With UKIDSS, we can use the information from [Hewett et al., 2006](https://ui.adsabs.harvard.edu/abs/2006MNRAS.367..454H/abstract)

|Band | magnitude offset (Δm)|
|-----|---------------------:|
|J    |                 0.938|
|H    |                 1.379|
|K    |                 1.900|

Convert Vega magnitudes to `AB`.

`````python
vega_cols     = ['W1_MAG', 'W2_MAG', 'JMAG', 'HMAG', 'KMAG']
vega_shift    = {'W1_MAG': 2.699, 'W2_MAG': 3.339, 'JMAG': 0.938, 'HMAG': 1.379, 'KMAG': 1.900}


for col in vega_cols:
    large_catalog_full_tab[col] += vega_shift[col]
`````

### Machine Learning

Machine Learning (ML) can help us derive trends, correlations, or predictions from datasets that would be impossible to analyse with traditional statistical methods.  

Also, and for that reason, ML is said to be just another application of statistics.

The main idea behind ML is that you can use a sub-set of your data to extract patterns and create models which can be applied to the rest of your data to predict the behaviour of certain properties. This approach can be much quicker than taking the full dataset and analyse each element one by one.

In our case, we are using magnitudes, which have similar ranges in their values. But, in general, it is better to re-scale all you quantities (features) to a common range.  

There are several ways of doing this (see https://scikit-learn.org/stable/auto_examples/preprocessing/plot_all_scaling.html#sphx-glr-auto-examples-preprocessing-plot-all-scaling-py or https://scikit-learn.org/stable/modules/preprocessing.html).  

We will use the `MinMaxScaler` from `scikit-learn`. It will make our data to be in the range $[0, 1]$

### Dimensionality Reduction

Now, we will try a simple method for understanding if there is any kind of *hidden* information in the data which might help us to derive a different property.

Dimensionality reduction methods are part of what is called Unsupervised Machine Learning.  

They are used as a way to create a *projection* of a full dataset (with tens, hundreds, or thousands of features) into a smaller table with fewer dimensions.  

For our example, we will take the full table and we will create a 2-D representation of it. With that, we might be able to see some trends on the positions of the elements of our catalogue.

There are several algorithms to reduce the dimensionality of a table, and we will use six of them.

- t-SNE
- UMAP
- PacMap
- Trimap
- SOM
- PCA

In order to run these methods, we need our table to be completely filled with values. Our table does not have that problem, so we can keep working.

Also, some of the methods use random number generation. So, and to be consistent, we will use the same random seed for all of them.

`````python
seed = 23
`````

Then, we can initialise the methods.

#### Using t-SNE

`````python
feat_red_tsne = TSNE(n_components=2, perplexity=20, metric='cosine', random_state=seed, n_jobs=-1)
`````

#### Using UMAP

`````python
feat_red_umap = umap_.UMAP(n_neighbors=30, n_components=2, random_state=seed)
`````

#### Using PacMap

`````python
feat_red_pacmap = pacmap.PaCMAP(n_dims=2, n_neighbors=None, MN_ratio=0.5, FP_ratio=2.0)#, random_state=seed)
`````

#### Using Trimap

`````python
feat_red_trimap = trimap.TRIMAP(n_inliers=20)
`````

#### Using SOM

`````python
feat_red_som = som.SOM(m=1, n=2, dim=12, random_state=seed)
`````

#### Using PCA

`````python
feat_red_pca = PCA(n_components=2)
`````

## Using Pycaret

Coding a full ML program which can take all our data, treat it, transform it, and create a model to predict some properties can be a tough task.  

In particular, in Astrophysics we are not necessarily experts in good coding practises and our programs can be very inefficient.  


That is why we might want to use ML solutions developed by experts. One of these solutions is **PyCaret** (https://pycaret.org). It can help us creating a full ML pipeline to analyse large datasets without loosing control over the parameters which create the models.

Please note that we still need to understand what PyCaret is doing under the hood in order to make sense of the result we might get from it.

PyCaret has many modules but we are going to use only its regression sub-module since we want to predict the redshift (a continuous variable) for our sources.

In general, in ML, people split their data in Training set, Testing set and Validation set (this might vary).  

- The training set is used to create the model. The data from these sample will give all their properties to predict new values.
- The testing set is used to check whether the new model is predicting good values. If that is the case, it will be included into the Training set and the model will be trained with this new set.
- The validation set is a part of the data that the model will never see. No information from this sub-set must be used to create (train) the model. Here is where we actually might know if we obtained a good model.

Consequently, we will separate the Validation set and PyCaret will take care of the rest. Note that we are only using the features used for the correlation calculations, but without any transformation. PyCaret will help us later with that.

`````python
data_train_test_df = corr_data_df.sample(frac=0.90, random_state=seed)  # Train + test sets
data_validation_df = corr_data_df.drop(data_train_test_df.index)  # Validation data

print('Shape of used data')
print('-' * 40)
print(f'Full Dataset size: {corr_data_df.shape}')
print(f'Data for Modeling (Train and Test): {data_train_test_df.shape}')
print(f'Unseen Data For Validation: {data_validation_df.shape}')
print('-' * 40)
`````

Now, we set PyCaret up. There are several options, and some of them might be different depending on whether we are running regression, classification, or other tasks (https://pycaret.readthedocs.io/en/latest/api/regression.html).

We will let PyCaret select the most relevant features by using the Boruta method (https://danielhomola.com/feature%20selection/phd/borutapy-an-all-relevant-feature-selection-method/).  

We might have also discarded the features with very high correlations, but we will not do it this time.

Also, we want to normalise our training features (not the target) with zscores (i.e., each feature will have a mean value, mu=0, and a standard deviation, sigma=1).

There are many more options (https://pycaret.readthedocs.io/en/latest/api/regression.html#pycaret.regression.setup).

Now, let's create our PyCaret experiment.

At this point, we define a specific way to check if the predictions we will make are correct (a metric).  

For redshift values, it will be:  

$$
\Delta z = \mathrm{median}\left(\frac{|(z_{True} - z_{Predict})|}{(1 + z_{True})}\right)
$$

`````python
def delta_z(y, y_pred, **kwargs):
    num = np.abs(y - y_pred)
    den = 1 + y
    return np.nanmedian(num / den)
`````

And we can add it to PyCaret to use it.

`````python
pyr.add_metric('z_met', 'delta z', delta_z, greater_is_better=False)
`````
